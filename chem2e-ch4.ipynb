#### Ex 4.77

The internet says gold is currently $5,053.11 / troy ounce.  The problem states

> Would you agree to buy 1 trillion (1,000,000,000,000) gold atoms for $5? Explain why or why not. ... (1 troy ounce=31.1 g)

```python
AvogadrosNumber = 6.022 * 10**23
10**12 # atoms
10**12 / AvogadrosNumber # moles
10**12 / AvogadrosNumber * b.Au # grams
10**12 / AvogadrosNumber * b.Au / 31.1 # troy ounces
10**12 / AvogadrosNumber * b.Au / 31.1 * 5053.11 # dollars
```

Answer: Yes, I would by 1 trillion gold atoms for 5 dollars, if I could use a microscope good enough to verify they were present.  However, that would just be for a conversation piece so I could say I owned 1 trillion gold atoms.  It would not be close to worth it as an investment, being worth a tiny fraction of a penny.
