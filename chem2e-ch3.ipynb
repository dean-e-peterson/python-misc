```python
def ex3_3(s):
    return {
        '3a': s.P*4,
        '3b': s.H*2 + s.O,
        '3c': s.Ca + (s.N + s.O*3)*2,
        '3d': s.C + s.H*3 + s.C + s.O*2 + s.H,
        '3e': s.C*12 + s.H*22 + s.O*11, }
compares(ex3_3)
```

```python
def ex3_4(s):
    return {
        '4a': s.C + s.Cl*2 + s.O,
        '4b': s.C*2 + s.H*2,
        '4c': s.C*2 + s.H*2 + s.Br*2,
        '4d': s.S + s.O*4 + s.H*2 }
compares(ex3_4)
```

```python
def ex3_5(s):
    return {
        '5a': s.C*2 + s.H*3 + s.C + s.H*2 + s.C + s.H*3,
        '5b': s.C*4 + s.H*6,
        '5c': s.Si*2 + s.Cl*4 + s.H*2,
        '5d': s.P*1 + s.O*4 + s.H*3 }
compares(ex3_5)
```

```python
def ex3_13(s):
    return {
        '13a': s.S*8,
        '13b': s.C*5 + s.H*12,
        '13c': s.Sc*2 + (s.S + s.O*4)*3,
        '13d': s.C + s.H*3 + s.C + s.O + s.C + s.H*3,
        '13e': s.C*6 + s.H*12 + s.O*6 }
compares(ex3_13)
```

```python
def ex3_15(s):
    return {
        '15a': s.C*2 + s.H + s.Br + s.Cl + s.F*3,
        '15b': s.C*12 + s.H*14 + s.N*2 + s.Cl*2,
        '15c': s.C*8 + s.H*10 + s.N*4 + s.O*2,
        '15d': s.C + s.O + (s.N + s.H*2)*2,
        '15e': s.C*17 + s.H*35 + s.C + s.O*2 + s.Na }
compares(ex3_15)
```

```python
def ex3_17(s):
    return {
        '17a': Decimal('0.0146') * (s.K + s.O + s.H),
        '17b': Decimal('10.2') * (s.C*2 + s.H*6),
        '17c': Decimal('0.0016') * (s.Na*2 + s.S + s.O*4),
        '17d': Decimal('6854') * (s.C*6 + s.H*12 + s.O*6),
        '17e': Decimal('2.86') * (s.Co + (s.N + s.H*3)*6 + s.Cl*3) }
compares(ex3_17)
```
